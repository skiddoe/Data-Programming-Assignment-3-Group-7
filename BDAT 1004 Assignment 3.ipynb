```python
count=0
total=0
largest=None
smallest=None
try:
    while True:
        number=input("Enter the number:  ")
        if number=='done':
            break
        print(number)
        count=count+1
        total=total+int(number)
        for intgr in number:
            if largest is None or intgr>largest:
                largest=intgr
            if smallest is None or intgr<smallest:
                smallest=intgr
    print("The total count is: ",count)
    print("Total of all number is:",total)
    print("Largest numbers is: ",largest )
    print("Smallest numbers is: ",smallest )
    
except:
    print("enter a valid number")
```

```python
dictionary_hours=dict()
least=[]
suum=0
kount=0
fname=input("Enter the file name:")
fhand=open(fname)
for line in fhand:
    line=line.rstrip()
    if line.startswith('From') and not line.startswith('From:'):
        words=line.split()
        hour=words[5][0:words[5].find(':')]
        least.append(hour)
for hour in least:
    if hour not in dictionary_hours:
         dictionary_hours[hour]=1
    else:
        dictionary_hours[hour]+=1
l=list()
for h,qty in list(dictionary_hours.items()):
    l.append((h,qty))
l.sort()
for h,qty in l:
    print(h,qty)
```

### Excercise 3

```python
#import string
import re
dictionary_counts=dict()
fname=input("Enter the file name:")
for line in fname:
    line=line.rstrip()
    line=line.lower()
    line=re.findall('[a-z]',line)
    for letter in line:
        if letter not in dictionary_counts:
            dictionary_counts[letter]=1
        else:
            dictionary_counts[letter]+=1
l=list()
for letter, qty in list(dictionary_counts.items()):
    l.append((qty,letter))
l.sort(reverse=True)
for letter, qty in l:
    print(qty,letter)
```
